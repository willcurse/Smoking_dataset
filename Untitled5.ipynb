# 3.1 Trend Analysis: How has smoking prevalence changed over the years for different age groups?

```python
plt.figure(figsize=(10,6))
sns.lineplot(y='Smoking_Prevalence',x='Year',data=data,hue='Age_Group',marker='o')
plt.title('How has smoking prevalence changed over the years for different age groups')
plt.xlabel('Year')
plt.ylabel('Smoking_Prevalence')
plt.show()
```

# 3.2 Age vs. Smoking: Which age group shows the highest smoking prevalence across all years?

```python
data.columns
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Age_Group',y='Smoking_Prevalence',data=data)
plt.title('Which age group shows the highest smoking prevalence across all years')
plt.xlabel('Age_Group')
plt.ylabel('Smoking_Prevalence')
plt.show()
```

# 3.3 Gender Differences: How does smoking prevalence vary between males and females within different age groups?

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Smoking_Prevalence',y='Gender',data=data)
plt.title('How does smoking prevalence vary between males and females within different age groups')
plt.xlabel('Smoking_Prevalence')
plt.ylabel('Gender')
plt.show()
```

# 3.4 Socioeconomic Status Impact: What is the relationship between socioeconomic status and smoking prevalence?

```python
data.columns
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Socioeconomic_Status',y='Smoking_Prevalence',data=data)
plt.title(' What is the relationship between socioeconomic status and smoking prevalence?')
plt.xlabel('Socioeconomic_Status')
plt.ylabel('Smoking_Prevalence')
plt.show()
```

# 3.5 Peer Influence: Is there a correlation between peer influence scores and smoking prevalence?

```python
data.columns
```

```python
#calculate corelation
correlation=data['Peer_Influence'].corr(data['Smoking_Prevalence'])
correlation

plt.figure(figsize=(8,6))
sns.scatterplot(x='Peer_Influence', y='Smoking_Prevalence', data=data)
plt.title(f'Correlation between Peer Influence and Smoking Prevalence is: {correlation:.2f}')
plt.xlabel('Peer Influence Score')
plt.ylabel('Smoking Prevalence (%)')
plt.grid(True)
plt.show()
```
